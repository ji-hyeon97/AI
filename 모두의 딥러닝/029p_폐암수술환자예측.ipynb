```python
#딥러닝을 구현하는데 필요한 케라스 함수
from keras.models import Sequential
from keras.layers import Dense

# 필요한 라이브러리를 불러옵니다.
import numpy #수치계산
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.
seed = 0
numpy.random.seed(seed)
#tf.set_random_seed(seed) 텐서플로우 라이브러리에 set_random_seed는 없다고 나온다 교재에는 있음...

# 준비된 수술 환자 데이터를 불러들입니다.
Data_set = numpy.loadtxt("/content/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장합니다.
X = Data_set[:,0:17] #폐활량, 고통, 기침, 흡연, 천식, 호흡곤란, 등의 정보
Y = Data_set[:,17] #수술후 생존 결과 1->생존, 0->사망

# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분입니다).
model = Sequential()#딥러닝 구조를 한 층 한 층 쉽게 쌓아 올릴 수 있도록 함
model.add(Dense(30, input_dim=17, activation='relu')) #은닉층 30개, 입력층 17개, relu함수(다음 층으로 값을 넘기는 법)
model.add(Dense(1, activation='sigmoid'))#클래스의 개수가 1개 -> 시그모이드

# 딥러닝을 실행합니다.
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])#optimizer는 오차를 어떻게 줄일지 정함
model.fit(X, Y, epochs=30, batch_size=10)#batch_size:가중치를 10번 업데이트(Adam)


# 결과를 출력합니다.
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
print((X,Y)[1])
```
