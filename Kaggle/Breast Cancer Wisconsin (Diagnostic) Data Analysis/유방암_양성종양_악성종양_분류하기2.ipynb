```python
import pandas as pd
df = pd.read_csv("data.csv")
df
```

```python
df.drop(['id'],axis=1, inplace=True)
df
#id값 필요 X
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df
#mapping
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df, test_size=0.3)
```

```python
import seaborn as sns
corr = df.iloc[:,1:12].corr()
corr
display(sns.heatmap(corr,cmap="coolwarm"))
```

```python
features = ['radius_mean','texture_mean','compactness_mean','smoothness_mean']
#중복되는 특성들 사이에 1개씩 선택
```

```python
train_input = train[features]
train_target = train['diagnosis']
test_input = test[features]
test_target = test['diagnosis']
```

```python
from sklearn.linear_model import LogisticRegression
#성능측정모듈 불러들이기
from sklearn import metrics
model1 = LogisticRegression()
model1.fit(train_input,train_target)
print(model1.intercept_, model1.coef_)
```

```python
pred1 = model1.predict(test_input)#예측하기
print(metrics.confusion_matrix(test_target,pred1))#혼동행렬출력
print(metrics.accuracy_score(test_target,pred1))
#15개 틀림
```

```python
print(metrics.classification_report(test_target,pred1))#성능지표출력
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(train_input, train_target)
pred2 = model2.predict(test_input)
print(metrics.confusion_matrix(test_target,pred2))#혼동행렬 출력
print(metrics.accuracy_score(test_target,pred2))
print(metrics.classification_report(test_target,pred2))#성능지표출력
#12개 틀림
```

```python
from sklearn.tree import export_graphviz
#결정트리 시각화
import graphviz
#100개의 결정트리중의 첫번째(0th_ index) 선택
tree = model2.estimators_[0]
#시각화
display(graphviz.Source(export_graphviz(tree,feature_names = features)))
#참_왼쪽, 거짓_오른쪽
```

```python
#부분 의존도 함수 불러들이기
from sklearn.inspection import plot_partial_dependence
#그래프 모듈 불러들이기
import matplotlib.pyplot as plt
#부분의존도 그래프 그리기
plot_partial_dependence(model1,train_input,features)

#그래프 보기 좋게 설정
fig = plt.gcf()
fig.set_size_inches(12,8)
fig.suptitle('Partial dependence of malignancy probability (Logistic) on each features')
fig.subplots_adjust()
#radius_mean이 결과예측에 큰 영향을 미친다
#연습에서 baxplot 시각화 참고하면 좋음..
```
